```python
heart_path = 'C:/Users/Ritik Gangwar/OneDrive/Desktop/ML Projects/Hair Style Recommendation System/FaceShape DataSet/training_set/Heart/'
data=[]
for img_file in heart_files:
  image = cv2.imread(heart_path+img_file)
  if image is not None: # Check if image was loaded successfully
    image = cv2.resize(image,(224,224))
    image = np.array(image)
    data.append(image)
  else:
    print(f"Failed to load image: {heart_path+img_file}") # Print error message if image loading fails

oblong_path = 'C:/Users/Ritik Gangwar/OneDrive/Desktop/ML Projects/Hair Style Recommendation System/FaceShape DataSet/training_set/Oblong/'
for img_file in oblong_files:
  image = cv2.imread(oblong_path+img_file)
  if image is not None: # Check if image was loaded successfully
    image = cv2.resize(image,(224,224))
    image = np.array(image)
    data.append(image)
  else:
    print(f"Failed to load image: {oblong_path+img_file}") # Print error message if image loading fails

oval_path = 'C:/Users/Ritik Gangwar/OneDrive/Desktop/ML Projects/Hair Style Recommendation System/FaceShape DataSet/training_set/Oval/'
for img_file in oval_files:
  image = cv2.imread(oval_path+img_file)
  if image is not None: # Check if image was loaded successfully
    image = cv2.resize(image,(224,224))
    image = np.array(image)
    data.append(image)
  else:
    print(f"Failed to load image: {oval_path+img_file}") # Print error message if image loading fails

round_path = 'C:/Users/Ritik Gangwar/OneDrive/Desktop/ML Projects/Hair Style Recommendation System/FaceShape DataSet/training_set/Round/'
for img_file in round_files:
  image = cv2.imread(round_path+img_file)
  if image is not None: # Check if image was loaded successfully
    image = cv2.resize(image,(224,224))
    image = np.array(image)
    data.append(image)
  else:
    print(f"Failed to load image: {round_path+img_file}") # Print error message if image loading fails

square_path = 'C:/Users/Ritik Gangwar/OneDrive/Desktop/ML Projects/Hair Style Recommendation System/FaceShape DataSet/training_set/Square/'
for img_file in square_files:
  image = cv2.imread(square_path+img_file)
  if image is not None: # Check if image was loaded successfully
    image = cv2.resize(image,(224,224))
    image = np.array(image)
    data.append(image)
  else:
    print(f"Failed to load image: {square_path+img_file}") # Print error message if image loading fails
```

```python
# Converting the image list and label list to numpy arrays
X = np.array(data)
Y = np.array(labels)
```

```python
print(X.shape)
print(Y.shape)
```

```python
# Train Test Split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# Scaling the data
X_train_scaled = X_train/255
X_test_scaled = X_test/255
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
import numpy as np
from tensorflow import keras
from tensorflow.keras import regularizers

# Define number of classes
number_of_classes = 5

# Build the model
model = keras.Sequential([
    keras.layers.Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(224,224,3),
                        kernel_regularizer=regularizers.l1(0.01)),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Conv2D(64, kernel_size=(3,3), activation='relu',
                        kernel_regularizer=regularizers.l1(0.01)),
    keras.layers.MaxPooling2D(pool_size=(2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu',
                       kernel_regularizer=regularizers.l1(0.01)),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(number_of_classes, activation='softmax',
                       kernel_regularizer=regularizers.l1(0.01))
])

# Print the model summary
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Check labels
print(np.unique(Y_train))

# Ensure labels are in the correct range
if np.min(Y_train) < 0 or np.max(Y_train) >= number_of_classes:
    raise ValueError(f"Labels should be in the range [0, {number_of_classes - 1}]")
```

```python
# Train the model
history = model.fit(X_train_scaled, Y_train, validation_split=0.1, epochs=10)
```

```python
loss, accuracy = model.evaluate(X_test_scaled, Y_test)
print('Test Accuracy: ', accuracy)
```

```python
h = history

# plot the loss value
plt.plot(h.history['loss'], label='train loss')
plt.plot(h.history['val_loss'], label='val loss')
plt.legend()
plt.show()

# plot the accuracy value
plt.plot(h.history['accuracy'], label='train acc')
plt.plot(h.history['val_accuracy'], label='val acc')
plt.legend()
plt.show()
```
